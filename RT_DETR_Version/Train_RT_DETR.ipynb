### Step 3: Train Transformers (Anti-Crash Version)
Starting 100 epochs on the `rtdetr-l.pt` engine. The `amp=False` toggle forces strictly 32-bit mathematics, making it physically impossible for the GPU to mathematically crash (`NaN/Inf Loss`) during gradient calculations.

```
from ultralytics import RTDETR

model = RTDETR('rtdetr-l.pt')

# `amp=False` fixes gradient explosions.
# `batch=8` stabilizes the learning curve.
results = model.train(data='Dataset with annotation in YOLO format/data.yaml', epochs=100, imgsz=640, project='RT_DETR_Coral', name='run_stable', amp=False, batch=8)
```

### Step 4: Direct Download
This deep search mechanism finds exactly where Ultralytics saved your model and legally forces the browser to download it to your local computer.

```
import glob, time
from google.colab import files

print("\u23f3 Searching for completely trained stable model...")
best_weights = glob.glob('RT_DETR_Coral/run_stable*/weights/best.pt')

if best_weights:
    print(f"\n\u2705 Found trained model at: {best_weights[-1]}")
    print("\u23f3 Prompting web browser download...")
    print("\u26a0\ufe0f DO NOT CLOSE THIS WINDOW UNTIL YOUR BROWSER DOWNLOADS THE .PT FILE! \u26a0\ufe0f")
    files.download(best_weights[-1])
else:
    print("\u274c Standard search failed. Commencing deep-dive server search...")
    all_models = glob.glob('/content/**/best.pt', recursive=True)
    if all_models:
        print(f"\u2705 Deep search recovered the model at: {all_models[-1]}")
        files.download(all_models[-1])
    else:
        print("\u274c Deep search failed. The training must have been terminated early.")
```
