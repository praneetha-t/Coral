# 🐠 Kaggle Cloud: RT-DETR Training

This version of the notebook is exclusively designed for the Kaggle Cloud environment. It automatically locates your uploaded dataset from the Kaggle Input directory, trains 100 safe epochs on the Kaggle GPU, and deposits your final `.pt` file permanently in the Kaggle Output section.

### Step 2: Auto-Detect Kaggle Dataset
Kaggle automatically unzips datasets you add into a hidden internal folder. This script dynamically searches the server to mathematically lock onto your `data.yaml` target.

```
import glob
import os

yaml_paths = glob.glob('/kaggle/input/**/data.yaml', recursive=True)

if yaml_paths:
    yaml_target = yaml_paths[0]
    print(f"\u2705 Dataset Auto-Locked: {yaml_target}")
else:
    print("\u274c FATAL ERROR: Could not find data.yaml! Did you add your dataset to the Kaggle notebook via the right sidebar?")
```

### Step 3: Unleash the Training
Start calculating the 100 stable, anti-crash epochs directly into Kaggle's `/kaggle/working/` output directory.

```
from ultralytics import RTDETR
import glob

yaml_target = glob.glob('/kaggle/input/**/data.yaml', recursive=True)[0]

model = RTDETR('rtdetr-l.pt')
results = model.train(data=yaml_target, epochs=100, imgsz=640, project='/kaggle/working/RT_DETR_Coral', name='run_stable', amp=False, batch=8)
```

### Step 4: Finalize Output
Copies the final successful model to the absolute root directory so you can easily download it from the generic 'Output' tab later.

```
import shutil
import os

best_weights = '/kaggle/working/RT_DETR_Coral/run_stable/weights/best.pt'
safe_export = '/kaggle/working/rtdetr_best.pt'

if os.path.exists(best_weights):
    shutil.copy(best_weights, safe_export)
    print("\u2705 TRAINING SUCCESSFUL! Model safely exported.")
    print(f"You can now download \'{safe_export}\' from the Output panel on the right side of your screen!")
else:
    print("\u274c Model compilation failed. No weights were generated.")
```
